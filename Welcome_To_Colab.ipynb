```
#create a class called animal with attributes name, age and a method called eat
class animal:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def eat(self):
        print("i like is eating omena like fala fulani")
#inheriting functions in pyhton
class dog(animal):
  def eat(self):
    print("This is a chihuahua")
dog1=dog("chihuahua", 2)
print(dog1.name())

```
